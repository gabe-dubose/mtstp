### Supplemental Figure 1

The distribution of pupal duration before freezing for samples collected at the late pupa stage. Plots show that the distribution of time late pupa samples spent as pupae before freezing was the same across all treatments.

```python
data = pd.read_csv("../metadata/mtstp_sample_metadata.tsv", sep="\t")
```

```python
#subset late pupa sample data by plant and infection status
cur_inf_late_pupae = data.loc[(data['plant-diet'] == 'curassavica') & (data['infection-status'] == 'infected') & (data['developmental-stage'] == 'Late-pupa')]
cur_uninf_late_pupae = data.loc[(data['plant-diet'] == 'curassavica') & (data['infection-status'] == 'uninfected') & (data['developmental-stage'] == 'Late-pupa')]

inc_inf_late_pupae = data.loc[(data['plant-diet'] == 'incarnata') & (data['infection-status'] == 'infected') & (data['developmental-stage'] == 'Late-pupa')]
inc_uninf_late_pupae = data.loc[(data['plant-diet'] == 'incarnata') & (data['infection-status'] == 'uninfected') & (data['developmental-stage'] == 'Late-pupa')]

sns.set_style('whitegrid')

#plot distributions of pupal age when frozen
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2, 2, figsize=(7, 7))
fig.tight_layout(pad=4)

cilp = sns.histplot(data=cur_inf_late_pupae, x="days-pupa-when-frozen", ax=ax1, discrete=True, alpha=1, color="tab:orange")
cilp.set_xticks(range(6,9))
cilp.set_yticks(range(5))
ax1.set_xlabel("Days since pupation")
ax1.set_title("$\it{A. curassavica}$ + Infected", fontdict={'fontsize': 10})

culp = sns.histplot(data=cur_uninf_late_pupae, x="days-pupa-when-frozen", ax=ax2, discrete=True, alpha=1, color="tab:green")
culp.set_xticks(range(6,9))
culp.set_yticks(range(5))
ax2.set_xlabel("Days since pupation")
ax2.set_title("$\it{A. curassavica}$ + Uninfected", fontdict={'fontsize': 10})

iilp = sns.histplot(data=inc_inf_late_pupae, x="days-pupa-when-frozen", ax=ax3, discrete=True, alpha=1, color="tab:red")
iilp.set_xticks(range(6,9))
iilp.set_yticks(range(5))
ax3.set_xlabel("Days since pupation")
ax3.set_title("$\it{A. incarnata}$ + Infected", fontdict={'fontsize': 10})

iulp = sns.histplot(data=inc_uninf_late_pupae, x="days-pupa-when-frozen", ax=ax4, discrete=True, alpha=1, color="tab:blue")
iulp.set_xticks(range(6,9))
iulp.set_yticks(range(5))
ax4.set_xlabel("Days since pupation")
ax4.set_title("$\it{A. incarnata}$ + Uninfected", fontdict={'fontsize': 10})

plt.savefig("late_pupal_age_distribution.pdf", bbox_inches ="tight")
```

```python
#import packages for analysis
import pandas as pd
from gnat import misc_utils
from gnat import error_calculator
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# load data and analysis metadata
data = pd.read_csv('data/counts_tables/dpl_tpm_counts_kallisto.csv')
data.rename(columns={data.columns[0]: "sample-id" }, inplace = True)
data.set_index("sample-id", inplace = True)
metadata = pd.read_csv('data/mtstp_analysis_metadata.tsv', sep="\t")

# instead of calculating all pairwise RMSE comparisons for all comparisons, 
# I'll make a function that for selecting specific developmental staes and 
# only data of interest for this analysis
def get_rmse_data(data, metadata, developmental_stage):
    #get stage data
    dev_metadata = metadata.loc[metadata["developmental-stage"] == developmental_stage]
    dev_ids = list(dev_metadata.loc[:, 'sample-id'])
    dev_data = data.loc[dev_ids]
    
    #get comparisons list
    comparisons = misc_utils.get_pairwise_comparisons(dev_data)
    rmse_results = error_calculator.pairwise_rmse(dev_data, comparisons)
    
    #add comparisons of relevance
    plant_comparison = []
    infection_comparison = []
    developmental_stage_comparison = []

    comparison1_values = list(rmse_results["comparison1"])
    comparison2_values = list(rmse_results["comparison2"])

    for i in range(len(comparison1_values)):
        sample1_id = comparison1_values[i]
        sample2_id = comparison2_values[i]

        #uninfected comparisons
        if "cu" in sample1_id and "iu" in sample2_id:
            plant_comparison.append("curassavica:incarnata")
            infection_comparison.append("uninfected:uninfected")
        elif "cu" in sample1_id and "cu" in sample2_id:
            plant_comparison.append("curassavica:curassavica")
            infection_comparison.append("uninfected:uninfected")
        elif "iu" in sample1_id and "iu" in sample2_id:
            plant_comparison.append("incarnata:incarnata")
            infection_comparison.append("uninfected:uninfected")
            
        #infected comparisons
        elif "ci" in sample1_id and "ii" in sample2_id:
            plant_comparison.append("curassavica:incarnata")
            infection_comparison.append("infected:infected")
        elif "ci" in sample1_id and "ci" in sample2_id:
            plant_comparison.append("curassavica:curassavica")
            infection_comparison.append("infected:infected")
        elif "ii" in sample1_id and "ii" in sample2_id:
            plant_comparison.append("incarnata:incarnata")
            infection_comparison.append("infected:infected")

        #if not a designated comparison, add "NA"
        else:
            plant_comparison.append("NA")
            infection_comparison.append("NA")
            
        #add developmental stage
        developmental_stage_comparison.append(developmental_stage)

    #add columns to dataframe
    rmse_results["infection_comparison"] = infection_comparison
    rmse_results["plant_comparison"] = plant_comparison
    rmse_results["developmental_stage"] = developmental_stage_comparison

    #select only columns of relevant comparisons
    rmse_comparison_data = rmse_results.loc[(rmse_results["infection_comparison"] != "NA") & (rmse_results["plant_comparison"] != "NA")]
    
    return rmse_comparison_data

#Get RMSE results for each developmental stage
thrid_instar_rmse_results = get_rmse_data(data, metadata, "third-instar")
print("Completed third instar RMSE comparisons")
fifth_instar_rmse_results = get_rmse_data(data, metadata, "fifth-instar")
print("Completed fifth instar RMSE comparisons")
early_pupa_rmse_results = get_rmse_data(data, metadata, "early-pupa")
print("Completed early pupa RMSE comparisons")
late_pupa_rmse_results = get_rmse_data(data, metadata, "late-pupa")
print("Completed late pupa RMSE comparisons")
adult_rmse_results = get_rmse_data(data, metadata, "adult")
print("Completed adult RMSE comparisons")

sns.set_style("darkgrid")
#Concatenate dataframes to make plotting easier
total_rmse_results = pd.concat([thrid_instar_rmse_results, fifth_instar_rmse_results, early_pupa_rmse_results, late_pupa_rmse_results, adult_rmse_results])
#select only incarnata vs curassavica comparisons
plant_rmse_results = total_rmse_results.loc[total_rmse_results["plant_comparison"] == "curassavica:incarnata"]
#make point plot
ax = sns.pointplot(data=plant_rmse_results, y="rmse", x="developmental_stage", hue="infection_comparison")
#tidy up plot a bit
ax_legend=ax.legend(prop=fm.FontProperties(size=12))
ax_legend.get_texts()[0].set_text('Infected')
ax_legend.get_texts()[1].set_text('Uninfected')
ax.set_ylabel('RMSE', size=12)
ax.set_xlabel('Developmental Stage', size=12)
```

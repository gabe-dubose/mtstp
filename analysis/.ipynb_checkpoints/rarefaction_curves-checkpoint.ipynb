```python
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Load data

```python
data = pd.read_csv('data/counts_tables/dpl_raw_counts_kallisto.csv')
data.rename(columns={data.columns[0]: "sample-id" }, inplace = True)
data.set_index("sample-id", inplace = True)
```

```python
data
```

```python
def assemble_gene_pool(data, sample_id):
    #get gene ids
    genes = list(data.columns)
    #get sample data
    sample_data = data.loc[sample_id]
    #assemble gene pool
    gene_pool = []
    for gene in genes:
        gene_count = int(sample_data[gene])
        for i in range(gene_count):
            gene_pool.append(gene)
    return gene_pool
    
```

```python
def rarefy(pool, minimum_sample, maximum_sample, step, dataframe, sample_id):
    #initialize list to hold unique gene counts
    gene_counts = []
    for i in range(minimum_sample, maximum_sample, step):
        #select genes
        gene_subset = random.sample(pool, i)
        #get number of unique genes selected
        unique_gene_count = len(set(gene_subset))
        gene_counts.append(unique_gene_count)
    #add gene counts to dataframe
    dataframe[sample_id] = gene_counts
        
```

#### Get rarefaction curves

```python
#define rarefaction curve parameters
minimum_sample = 100000
maximum_sample = 2005000
step = 50000

#initialize rarefaction data frame
rarefaction_curves = pd.DataFrame()

#add subsample sizes to dataframe
subsample_size = []
for i in range(minimum_sample, maximum_sample, step):
    subsample_size.append(i)
rarefaction_curves["subsample-size"] = subsample_size

#for each sample in input dataframe, get rarefaction curve
for sample_id in list(data.index):
    #assemble gene pool
    gene_pool = assemble_gene_pool(data, sample_id)
    #get gene count at designated depths
    rarefy(gene_pool, minimum_sample, maximum_sample, step, rarefaction_curves, sample_id)
    print(f"Completed rarefaction curve for {sample_id}")
```

```python
rarefaction_curves
```

#### Make rarefaction curve plots

```python
sns.set_style("darkgrid")
ax = sns.lineplot()
for sample in list(rarefaction_curves.columns)[1:]:
    sns.lineplot(data=rarefaction_curves, x="subsample-size", y=sample, ax=ax, color="tab:blue", alpha=0.5)
ax.set_ylabel("Number of Genes Detected")
ax.set_xlabel("Sequencing Depth")

plt.savefig("../figures/rarefaction.pdf")
```

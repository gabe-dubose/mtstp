```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read metabolism and organismal systems data
metabolism_df = pd.read_csv('data/function_tables/metabolism_nonredundant.csv')
organismal_df = pd.read_csv('data/function_tables/organismal_systems_nonredundant.csv')
data = pd.merge(metabolism_df, organismal_df, on='sample-id')
```

### Exponential model

```python
#it looks like energy metabolism and immunity are pretty heavily negatively correlated
#I'll try to plot these directly

samples = list(organismal_df.index)
immunity_metabolism_data = pd.DataFrame(index=samples)

plant = list(organismal_df["plant"])
stage = list(organismal_df["developmental_stage"])
immune_expression = list(organismal_df["Immune system"])
energy_metabolism = list(metabolism_df["Energy metabolism"])
carbohydrate_metabolism = list(metabolism_df["Carbohydrate metabolism"])

immunity_metabolism_data['Immunity'] = immune_expression
immunity_metabolism_data['Energy_metabolism'] = energy_metabolism
immunity_metabolism_data['Carbohydrate_metabolism'] = carbohydrate_metabolism

immunity_metabolism_data['stage']= stage
immunity_metabolism_data['plant']= plant


#divide everything by 10,000. This makes the unit transceipts per hundred, so they read as percentages
#this is also necessary to exponentiate
immunity_metabolism_data['Immunity'] = immunity_metabolism_data['Immunity'] / 10000
immunity_metabolism_data['Energy_metabolism'] = immunity_metabolism_data['Energy_metabolism'] / 10000
immunity_metabolism_data['Carbohydrate_metabolism'] = immunity_metabolism_data['Carbohydrate_metabolism'] / 10000
```

```python
immunity_metabolism_data
```

```python
# the exponential decay function
def exponential_decay(x, a, b, c):
    return a * np.exp(-b * x) + c

#a function to generate predictions and CI
def fit_exponential_decay(data, x_col, y_col):
    x = np.array(list(data[x_col]))
    y = np.array(list(data[y_col]))
    
    params, covariance = curve_fit(exponential_decay, x, y)
    a, b, c = params
    
    # Extract the fitted parameters and their standard errors
    sigma_a, sigma_b, sigma_c = np.sqrt(np.diag(covariance))
    
    # Generate the fitted curve using the parameters
    y_fit = exponential_decay(x, a, b, c)
    
    # Calculate the residuals
    residuals = y - y_fit
    
    x_obs, y_pred = zip(*sorted(zip(x, y_fit)))
    ci = 0.1 * np.std(y_pred) / np.mean(y_pred)
    
    output = {'y_pred' : y_pred, 'residuals' : residuals, 'x_values' : x_obs, 'ci' : ci}
    
    print("\nExponential Decay Model Fit Summary:\n")
    print(f"Fitted Parameters:\n"
      f"a = {a} (SE = {sigma_a})\n"
      f"b = {b} (SE = {sigma_b})\n"
      f"c = {c} (SE = {sigma_c})")

    sum_squared_residuals = np.sum(residuals**2)
    print(f"Sum of squared residuals: {sum_squared_residuals}")
    
    return output
```

```python
#immunty vs energy metabolism model
immunity_vs_energy_exp_decay_model = fit_exponential_decay(data = immunity_metabolism_data, 
                                                x_col = 'Energy_metabolism', 
                                                y_col = 'Immunity')
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(10, 5))

#plot immunity vs energy metabolism
sns.scatterplot(data=immunity_metabolism_data, y="Immunity", x="Energy_metabolism", hue="stage", ax=ax1)
sns.lineplot(x=immunity_vs_energy_exp_decay_model['x_values'], 
             y=immunity_vs_energy_exp_decay_model['y_pred'], ax=ax1)
ax1.fill_between(immunity_vs_energy_exp_decay_model['x_values'], 
                 (immunity_vs_energy_exp_decay_model['y_pred']-immunity_vs_energy_exp_decay_model['ci']), 
                 (immunity_vs_energy_exp_decay_model['y_pred']+immunity_vs_energy_exp_decay_model['ci']), 
                 color='tab:blue', alpha=0.15)
ax1.set_xlabel('Energy Metabolism (TPH)')
ax1.set_ylabel('Immunity (TPH)')
```

```python
import numpy as np
from scipy.optimize import curve_fit

# The exponential decay function as a differential equation
def exponential_decay_diff_eq(x, y0, k, c):
    return y0 * np.exp(-k * x) + c

# Function to fit the exponential decay differential equation to data
def fit_exponential_decay(data, x_col, y_col):
    x = np.array(list(data[x_col]))
    y = np.array(list(data[y_col]))

    params, covariance = curve_fit(exponential_decay_diff_eq, x, y)
    y0, k, c = params

    # Extract the standard errors of the fitted parameters
    sigma_y0, sigma_k, sigma_c = np.sqrt(np.diag(covariance))

    # Generate the fitted curve using the parameters
    y_fit = exponential_decay_diff_eq(x, y0, k, c)

    # Calculate the residuals
    residuals = y - y_fit

    x_obs, y_pred = zip(*sorted(zip(x, y_fit)))
    ci = 0.1 * np.std(y_pred) / np.mean(y_pred)

    output = {'y_pred': y_pred, 'residuals': residuals, 'x_values': x_obs, 'ci': ci}

    print("\nExponential Decay Model Fit Summary:\n")
    print(f"Fitted Parameters:\n"
          f"y0 = {y0} (SE = {sigma_y0})\n"
          f"k = {k} (SE = {sigma_k})\n"
          f"c = {c} (SE = {sigma_c})")

    sum_squared_residuals = np.sum(residuals ** 2)
    print(f"Sum of squared residuals: {sum_squared_residuals}")

    return output
```

### MIT Model

##### Deriving model

Because the exponential decay model fit so well, I'll start with trying to make the model in this shape, but with biologically significant features. 

I'll start with:

1) <i>dI/dM = I*T</i>
     
where I is immune investment and T is the strength of trade off between immunity and metabolism investment. In biological terms, I think this would be how much transcriptional investment that could have been placed in immunity is diverted to metabolism.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the differential equation for dI/dM = I * T
def model(I, M, T):
    dIdM = I * T
    return dIdM

# Define the initial conditions for I
I0 = 1.0  # Initial value of I

# Define the range of values for M
M_values = np.linspace(0, 10, 100)  # Values of M

# Different values of T to iterate through
T_values = [-0.2, -0.4, -0.6, -0.8, -1]  # You can adjust or add more values

# Plot both results for different T values
plt.figure(figsize=(12, 6))

# Plot of dI/dM = I * T for different T values
plt.subplot(1, 2, 1)
for T in T_values:
    I_solution = [I0]
    for M in M_values[1:]:
        dM = M_values[1] - M_values[0]
        dI = model(I_solution[-1], M, T) * dM
        I_solution.append(I_solution[-1] + dI)
    plt.plot(M_values, I_solution, label=f'T = {T}')

plt.xlabel('M')
plt.ylabel('I')
plt.title('Differential Equation Solution (dI/dM = I * T)')
plt.legend()

# Plot of different T values
plt.subplot(1, 2, 2)
for T in T_values:
    T_solution = [T] * len(M_values)
    plt.plot(M_values, T_solution, label=f'T = {T}')

plt.xlabel('M')
plt.ylabel('T')
plt.title('Different T values as a function of M')
plt.legend()

plt.tight_layout()
plt.show()
```

So this model assumes that the rate at which transcription is diverted from immunity to metabolsim is constant with respect to metabolic investment. However, I think that if there is increased metabolic investment, increases the rate at which transcription is diverted from immunity towards metabolism. In other words, <i>T</i> becomes more negative. To accomplish this, I'll multiply <i>T</i> by metabolic investment <i>M</i>. This will scale the effects <i>T</i> to how much transcription is being invested in metabolism.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the differential equation for dI/dM = I * T * M
def model(I, M, T):
    dIdM = I * T * M
    return dIdM

# Define the initial condition for I
I0 = 1.0  # Initial value of I
M_values = np.linspace(0, 10, 100)  # Values of M over which you want to solve the equation

# Define different values of T to iterate through
T_values = [-0.2, -0.4, -0.6, -0.8, -1]  # You can adjust or add more values

# Plot both results side by side
plt.figure(figsize=(12, 6))

# Plot of dI/dM = I * T * M for different T values
plt.subplot(1, 2, 1)
for T in T_values:
    I_solution = odeint(model, I0, M_values, args=(T,))
    plt.plot(M_values, I_solution, label=f'T = {T}')

plt.xlabel('M')
plt.ylabel('I')
plt.title('dI/dM = I * T * M')
plt.legend()

# Plot of T * M for different T values
plt.subplot(1, 2, 2)
for T in T_values:
    T_M_values = T * M_values
    plt.plot(M_values, T_M_values, label=f'T = {T}')

plt.xlabel('M')
plt.ylabel('T')
plt.title('T = T*M')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit

# The exponential decay function as a differential equation
def exponential_decay_diff_eq(x, y0, k):
    return y0*np.exp(k*(x-x))

# Function to fit the exponential decay differential equation to data
def fit_exponential_decay(data, x_col, y_col):
    x = np.array(list(data[x_col]))
    y = np.array(list(data[y_col]))

    params, covariance = curve_fit(exponential_decay_diff_eq, x, y)
    y0, k = params

    # Extract the standard errors of the fitted parameters
    sigma_y0, sigma_k = np.sqrt(np.diag(covariance))

    # Generate the fitted curve using the parameters
    y_fit = exponential_decay_diff_eq(x, y0, k)

    # Calculate the residuals
    residuals = y - y_fit

    x_obs, y_pred = zip(*sorted(zip(x, y_fit)))
    ci = 0.1 * np.std(y_pred) / np.mean(y_pred)

    output = {'y_pred': y_pred, 'residuals': residuals, 'x_values': x_obs, 'ci': ci}

    print("\nExponential Decay Model Fit Summary:\n")
    print(f"Fitted Parameters:\n"
          f"y0 = {y0} (SE = {sigma_y0})\n"
          f"k = {k} (SE = {sigma_k})\n")

    sum_squared_residuals = np.sum(residuals ** 2)
    print(f"Sum of squared residuals: {sum_squared_residuals}")

    return output
```

Now <i>T</i> is scaled with <i>M</i>. While this is probably quite a bit better than assuming <i>T</i> is constant, I want to incorporate a little more flexibility. Looking at my actual data, it seems that third instar caterpillars are investing quite a bit more in immunity than predicted from their investment in metabolism. What this means is that there must be more investment in both immunity and metabolism together, which would mean there is less investment in other processes. Therefore, it is possible that investing more in both metabolism and immunity lightens the strength of trade off between the two.

I think this makes sense. Say that an individual must invest X amount of energy into metabolism to grow optimally. Then say there is additional investment in the combination of metabolsim and immunity, which takes away from investment in other processes. This could theoretically lighten the trade off between metabolism and immunity (but it increases the trade off between immunity+metabolism and other processes). 

Building in this dynamic could be done a couple of ways. One option is to include a term that scales the strength of trade off <i>T</i> with the total investment in immunity and metabolism (<i>M+I</i>). Alternatively, we could include an additional parameter that defines the minimum metabolic investment needed for growth. Then I could say that any excess investment in immunity+metabolsim goes towards immunity when the minimal metabolic investment is met. I don't know how much I like this idea because I'm trying to build in parameters that I can estimate from my data, and estimating minimal metabolic investment sounds difficult in general. The minimal metabolic investment likely changes throughout ontogeny as well. Maybe I'll put a pin in this and just go with scaling <i>T</i> to the <i>M+T</i> investment. Overall, increasing the investment in immunity+metabolism should decrease <i>T</i>.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the differential equation
def model(I, M, T, E):
    return I * (T * M / E)

# Given vectors of I and M values
# Replace these with your actual I and M values
I_values = np.linspace(1, 5, 100)  # Example values; replace with your actual I values
M_values = np.linspace(0.1, 10, 100)  # Example values; replace with your actual M values

# Calculate E values (sum of I and M)
E_values = I_values + M_values

# Values of T to iterate through
T_values = [-0.2, -0.4, -0.6, -0.8, -1]  # Change the range and number of values as needed

plt.figure(figsize=(12, 6))

# Plot for dI/dM = I * (T * M / E) for different values of T
plt.subplot(1, 2, 1)
for T in T_values:
    solutions = []
    for I, M, E in zip(I_values, M_values, E_values):
        I_solution = 1  # Initial condition for I
        I_solution += model(I_solution, M, T, E) * (M_values[1] - M_values[0])
        solutions.append(I_solution)
    plt.plot(M_values, solutions, label=f'T = {T}')
plt.xlabel('M')
plt.ylabel('I')
plt.title('dI/dM = I * (T * M / E)')
plt.legend()

# Plot for T * M / E with respect to M
plt.subplot(1, 2, 2)
for T in T_values:
    TM_over_E = T * M_values / E_values
    plt.plot(M_values, TM_over_E, label=f'T = {T}')
plt.xlabel('M')
plt.ylabel('T * M / E')
plt.title('T * M / E')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from gnat import rarefaction
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Load data

```python
data = pd.read_csv('data/counts_tables/dpl_raw_counts_kallisto.csv')
data.rename(columns={data.columns[0]: "sample-id" }, inplace = True)
data.set_index("sample-id", inplace = True)
```

```python
data
```

#### Get rarefaction curves

```python
parameters = {'data' : data, 'minimum_sample' : 1000000, 'maximum_sample': 'total', 'step' : 1000000}
samples = list(data.index)
rarefaction_curves = rarefaction.generate_rarefaction_curves(samples, parameters)
```

#### Plot rarefaction curves

```python
sns.set_style("darkgrid")

#initialize plot
ax = sns.lineplot()

#for each sample, add curve to plot
for curve in rarefaction_curves:
    for sample in curve:
        counts = curve[sample]['count']
        depths = curve[sample]['depth']
        sns.lineplot(x=depths, y=counts, ax=ax, color="tab:blue", alpha=0.5)
ax.set_ylabel("Number of Genes Detected")
ax.set_xlabel("Sequencing Depth")
plt.savefig('../figures/rarefaction.pdf')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Time from hatching to pupation
Statistically significant differences between plant species but not infection status

```python
data = pd.read_csv("../metadata/mtstp_sample_metadata.tsv", sep="\t")
```

```python
time_until_pupation_data = data[['sample-id', 'developmental-stage', 'plant-diet', 'infection-status', 'days-hatch-to-pupation']].dropna(subset="days-hatch-to-pupation")

sns.set_style('whitegrid')
plot = sns.histplot(data=time_until_pupation_data, x="days-hatch-to-pupation", hue="plant-diet", discrete=True, alpha=0.5)
plot.set_xticks(range(10,14))
plot.set_xlabel("Time from hatching to pupation (days)")
```

### Time from pupation to eclosion
ANOVA was statistically significantly different. Tukey HSD showed no pairwise differences. However, the differences between infected:incarnata and uninfected:incarnata was almost significant (p=0.0884416)

```python
data = pd.read_csv("../metadata/mtstp_sample_metadata.tsv", sep="\t")
```

```python
time_from_pupation_to_eclosion = data[['sample-id', 'developmental-stage', 'plant-diet', 'infection-status', 'days-pupation-to-eclosion']].dropna(subset="days-pupation-to-eclosion")

sns.set_style('whitegrid')
plot = sns.barplot(data=time_from_pupation_to_eclosion, x="plant-diet", y="days-pupation-to-eclosion", hue="infection-status")
plot.set(ylim=(6,10.5))
```

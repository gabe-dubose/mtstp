```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Time from hatching to pupation
Statistically significant differences between plant but not infection status

```python
data = pd.read_csv("../metadata/mtstp_sample_metadata.tsv", sep="\t")
```

```python
time_until_pupation_data = data[['sample-id', 'time-point', 'plant', 'infection-status', 'time-from-hatch-to-pupation-days']].dropna(subset="time-from-hatch-to-pupation-days")
```

```python
sns.set_style('whitegrid')
plot = sns.histplot(data=time_until_pupation_data, x="time-from-hatch-to-pupation-days", hue="plant", discrete=True, element="step", alpha=0.5)
plot.set_xticks(range(10,14))
plot.set_xlabel("Time from hatching to pupation (days)")
```

### Time pupation to eclosion
ANOVA was statistically significantly different. Tukey HSD showed no pairwise differences. However, the differences between infected:incarnata and uninfected:incarnata was almost significant (p=0.0884416)

```python
data = pd.read_csv("../metadata/mtstp_sample_metadata.tsv", sep="\t")
```

```python
time_from_pupation_to_eclosion = data[['sample-id', 'time-point', 'plant', 'infection-status', 'time-from-pupation-to-eclosion-days']].dropna(subset="time-from-pupation-to-eclosion-days")

sns.set_style('whitegrid')
plot = sns.barplot(data=time_from_pupation_to_eclosion, x="plant", y="time-from-pupation-to-eclosion-days", hue="infection-status")
plot.set(ylim=(6,10.5))
```

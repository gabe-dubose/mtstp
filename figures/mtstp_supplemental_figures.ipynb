```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Supplemental Figure 1

The distribution of pupal duration before freezing for samples collected at the late pupa stage. Plots show that the distribution of time late pupa samples spent as pupae before freezing was the same across all treatments.

```python
data = pd.read_csv("../metadata/mtstp_sample_metadata.tsv", sep="\t")
```

```python
#subset late pupa sample data by plant and infection status
cur_inf_late_pupae = data.loc[(data['plant'] == 'curassavica') & (data['infection-status'] == 'infected') & (data['time-point'] == 'Late-pupa')]
cur_uninf_late_pupae = data.loc[(data['plant'] == 'curassavica') & (data['infection-status'] == 'uninfected') & (data['time-point'] == 'Late-pupa')]

inc_inf_late_pupae = data.loc[(data['plant'] == 'incarnata') & (data['infection-status'] == 'infected') & (data['time-point'] == 'Late-pupa')]
inc_uninf_late_pupae = data.loc[(data['plant'] == 'incarnata') & (data['infection-status'] == 'uninfected') & (data['time-point'] == 'Late-pupa')]

sns.set_style('whitegrid')

#plot distributions of pupal age when frozen
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2, 2, figsize=(7, 7))
fig.tight_layout(pad=4)

cilp = sns.histplot(data=cur_inf_late_pupae, x="pupal-age-when-frozen-days", ax=ax1, discrete=True, alpha=1, color="tab:orange")
cilp.set_xticks(range(6,9))
cilp.set_yticks(range(5))
ax1.set_xlabel("Days since pupation")
ax1.set_title("$\it{A. curassavica}$ + Infected", fontdict={'fontsize': 10})

culp = sns.histplot(data=cur_uninf_late_pupae, x="pupal-age-when-frozen-days", ax=ax2, discrete=True, alpha=1, color="tab:green")
culp.set_xticks(range(6,9))
culp.set_yticks(range(5))
ax2.set_xlabel("Days since pupation")
ax2.set_title("$\it{A. curassavica}$ + Uninfected", fontdict={'fontsize': 10})

iilp = sns.histplot(data=inc_inf_late_pupae, x="pupal-age-when-frozen-days", ax=ax3, discrete=True, alpha=1, color="tab:red")
iilp.set_xticks(range(6,9))
iilp.set_yticks(range(5))
ax3.set_xlabel("Days since pupation")
ax3.set_title("$\it{A. incarnata}$ + Infected", fontdict={'fontsize': 10})

iulp = sns.histplot(data=inc_uninf_late_pupae, x="pupal-age-when-frozen-days", ax=ax4, discrete=True, alpha=1, color="tab:blue")
iulp.set_xticks(range(6,9))
iulp.set_yticks(range(5))
ax4.set_xlabel("Days since pupation")
ax4.set_title("$\it{A. incarnata}$ + Uninfected", fontdict={'fontsize': 10})
```
